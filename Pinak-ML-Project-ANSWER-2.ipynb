## Question 2.2)  Remove all the stopwords from the three speeches.


<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#0e92ea"> Answer 2.2:

### Reading the text file into a dataframe for further processing.....

```python
df_Roosevelt=pd.read_csv("Roosevelt.txt", delimiter = "\t", names = ["Roosevelt_speech"])
df_Roosevelt
```

```python
df_Kennedy=pd.read_csv("Kennedy.txt", delimiter = "\t", names = ["Kennedy_speech"])
df_Kennedy
```

```python
df_Nixon=pd.read_csv("Nixon.txt", delimiter = "\t", names = ["Nixon_speech"])
df_Nixon
```

### Converting to lower case : 

```python
df_Roosevelt['Roosevelt_speech'] = df_Roosevelt['Roosevelt_speech'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df_Roosevelt['Roosevelt_speech'].head()
```

```python
df_Roosevelt['Roosevelt_speech']= df_Roosevelt['Roosevelt_speech'].str.replace('"',"").str.replace('[^\w\s]','')
```

```python
df_Roosevelt['Roosevelt_speech'].head()
```

```python
df_Kennedy['Kennedy_speech'] = df_Kennedy['Kennedy_speech'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df_Kennedy['Kennedy_speech']= df_Kennedy['Kennedy_speech'].str.replace('"',"").str.replace('[^\w\s]','')
df_Kennedy['Kennedy_speech'].head()
```

```python
df_Nixon['Nixon_speech'] = df_Nixon['Nixon_speech'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df_Nixon['Nixon_speech']= df_Nixon['Nixon_speech'].str.replace('"',"").str.replace('[^\w\s]','')
df_Nixon['Nixon_speech'].head()
```

### Removing Stopwords with nltk.stopwords library

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
#stop_words=set(stopwords.words("english"))
stopwords = nltk.corpus.stopwords.words('english') +list(string.punctuation)
stopwords.extend(["--"])
```

```python
all_words_Roosevelt = [x for x in pd.Series(' '.join(df_Roosevelt['Roosevelt_speech']).split())] 
```

```python
Roosevelt_Without_Stopwords= [word for word in all_words_Roosevelt if word not in stopwords]
```

```python
Roosevelt_Without_Stopwords
```

```python
all_words_Kennedy = [x for x in pd.Series(' '.join(df_Kennedy['Kennedy_speech']).split())] 
```

```python
Kennedy_Without_Stopwords= [word for word in all_words_Kennedy if word not in stopwords]
```

```python
Kennedy_Without_Stopwords
```

```python
all_words_Nixon = [x for x in pd.Series(' '.join(df_Nixon['Nixon_speech']).split())] 
```

```python
Nixon_Without_Stopwords= [word for word in all_words_Nixon if word not in stopwords]
```

```python
Nixon_Without_Stopwords
```

## Question 2.3)  Which word occurs the most number of times in his inaugural address for each president? Mention the top three words.


<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#0e92ea"> Answer 2.3:

### Using nltk.FreqDist function to find the most occurance word.

```python
nltk.FreqDist(Roosevelt_Without_Stopwords).most_common(5)
```

```python
nltk.FreqDist(Kennedy_Without_Stopwords).most_common(5)
```

```python
nltk.FreqDist(Nixon_Without_Stopwords).most_common(5)
```

## We have seen that the most occured words in Kennedy and Nixon is "let" and "us". "let" is a generic word, while "us" may have been referred to "Americans". With that assumption, "let" can be removed form the most frequent word.

### So, top three words:
- Roosevelt: Nation, Know, Spirit
- Kennedy: Us, World, Sides
- Nixon: Us, Peace, world 

## Question 2.4)  Plot the word cloud of each of the speeches of the variable. (after removing the stopwords)


<span style="font-family: Arial; font-weight:bold;font-size:1.5em;color:#0e92ea"> Answer 2.4:

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
wordcloud_ROOS = WordCloud(stopwords=STOPWORDS,
                      width=3000,
                      height=2500
                     ).generate(Roosevelt)
```

```python
plt.figure(figsize=(12, 12))
plt.imshow(wordcloud_ROOS)
plt.axis('off')
plt.show()
```
